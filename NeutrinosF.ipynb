$$
\mu_{90} = T \, \Delta\Omega \int_{E_{\min}}^{E_{\max}}
\Phi_{90}(E) \, A_{\mathrm{eff}}(E) \, dE
$$

For a **bin-averaged** effective area, we approximate:  

$$
\mu_{90} \approx T \, \Delta\Omega \, \Phi_{90}(E_{\mathrm{geo}}) \, \Delta E \, \langle A_{\mathrm{eff}}\rangle
$$

where for **$\log_{10}$-spaced** bins:  

$$
\Delta E \approx E_{\mathrm{geo}} \, \ln 10 \, \Delta \log_{10} E,
\quad E_{\mathrm{geo}} = \sqrt{E_{\min}E_{\max}}
$$

If the sensitivity is expressed as **$E^{2}\Phi_{90}$**:  

$$
\Phi_{90}(E_{\mathrm{geo}}) =
\frac{\left(E^{2}\Phi_{90}\right)}{E_{\mathrm{geo}}^{2}}
$$
